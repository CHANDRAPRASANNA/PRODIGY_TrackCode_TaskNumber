```python
#Task 1

#Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("//content//drive//MyDrive//API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv", skiprows=4)
df_2022 = df[['Country Name', '2022']].dropna().sort_values(by='2022', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(df_2022['Country Name'], df_2022['2022'] / 1e6, color='teal')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Most Populous Countries in 2022')
plt.ylabel('Population (Millions)')
plt.xlabel('Country')
plt.tight_layout()
plt.show()
```
